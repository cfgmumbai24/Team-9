```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from pycaret.utils import version
from pycaret.regression import setup, compare_models, create_model
```

```python
print(version())
```

```python
df = pd.read_csv('/kaggle/input/dummy-student-success/dummy_student_success.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
s = setup(data= df, target='Student Success Rate', normalize= True)
```

```python
cm = compare_models()
```

```python
# model = create_model('lr')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop(['Student Success Rate','Student_ID'], axis = 1)
y = df['Student Success Rate']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', LinearRegression())
])
```

```python
pipe.fit(X_train,y_train)
```

```python
pred = pipe.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_test,pred))
```

```python
import joblib
```

```python
joblib.dump(pipe, 'student_success_rate.pkl')
```
